```python
#3.2 Data Preparation
```

```python
import pandas as pd

# Load the datasets
data_x = pd.read_csv('Data_X.csv')
data_y = pd.read_csv('Data_Y.csv')
data_new_x = pd.read_csv('DataNew_X.csv')

# Step 1: Merge datasets based on 'ID'
merged_data = pd.merge(data_x, data_y, on='ID', how='inner')

merged_data.head()
```

```python

# Step 2: Check for missing values
missing_values = merged_data.isnull().sum()
missing_values
```

```python
# Step 3: Quick overview of data types and comparability of attributes
data_info = merged_data.info()
data_info
```

```python
# Check if values of different attributes are comparable
# Example: checking if columns in Data X and DataNew X match
common_columns = set(data_x.columns).intersection(set(data_new_x.columns))
print("\nCommon columns between Data X and DataNew X:")
print(common_columns)
```

```python
# Step 4: Handling missing values (for now, we'll drop rows with missing values)
cleaned_data = merged_data.dropna()
cleaned_data
```

```python
# Step 5: Split data into training and test subsets
from sklearn.model_selection import train_test_split

# Use an 80/20 split for training and testing
train_data, test_data = train_test_split(cleaned_data, test_size=0.2, random_state=42)

# Output the results
train_data.shape, test_data.shape
```

```python
#3.4 Data Modeling
```

```python
# Removing outliers based on Z-score show in box plot
from scipy import stats
import numpy as np
z_scores = stats.zscore(train_data[['FR_COAL', 'DE_HYDRO', 'FR_HYDRO', 'DE_WINDPOW', 
                              'FR_WINDPOW', 'DE_LIGNITE', 'DE_RESIDUAL_LOAD', 
                              'FR_RESIDUAL_LOAD', 'DE_RAIN', 'FR_RAIN', 'DE_WIND', 
                              'FR_WIND', 'DE_TEMP', 'FR_TEMP', 'GAS_RET', 
                              'COAL_RET', 'CARBON_RET']])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
data_cleaned = train_data[filtered_entries]
```

```python
 #Apply one-hot encoding to the 'COUNTRY' column
data_encoded = pd.get_dummies(data_cleaned, columns=['COUNTRY'], drop_first=True)

# Verify the new data
print(data_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Re-select features and target after encoding
features = data_encoded.drop(['ID', 'TARGET'], axis=1)
target = data_encoded['TARGET']

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Verify the scaled features
print(features_scaled[:5])  # Print the first 5 rows for checking
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.3, random_state=42)

# Initialize and fit the model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict and evaluate
y_pred = lr.predict(X_test)
print("Linear Regression")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R^2:", r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)
print("Ridge Regression")
print("MSE:", mean_squared_error(y_test, y_pred_ridge))
print("R^2:", r2_score(y_test, y_pred_ridge))
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
print("Lasso Regression")
print("MSE:", mean_squared_error(y_test, y_pred_lasso))
print("R^2:", r2_score(y_test, y_pred_lasso))
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)
print("K-Nearest Neighbors Regression")
print("MSE:", mean_squared_error(y_test, y_pred_knn))
print("R^2:", r2_score(y_test, y_pred_knn))
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
print("Decision Tree Regression")
print("MSE:", mean_squared_error(y_test, y_pred_dt))
print("R^2:", r2_score(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
print("Random Forest Regression")
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("R^2:", r2_score(y_test, y_pred_rf))
```

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import spearmanr
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_squared_error as rmse
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'K-NN': KNeighborsRegressor(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor()
}

params = {
    'Ridge Regression': {'alpha': [0.1, 1, 10]},
    'Lasso Regression': {'alpha': [0.1, 1, 10]},
    'K-NN': {'n_neighbors': [5, 10, 15], 'p': [1, 2]},
    'Decision Tree': {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5]},
    'Random Forest': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
}
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error as mse, r2_score
from scipy.stats import spearmanr

results = {}
for name, model in models.items():
    if name in params:
        search = GridSearchCV(model, params[name], cv=5, scoring='neg_mean_squared_error')
        search.fit(features_scaled, target)
        best_model = search.best_estimator_
    else:
        best_model = model.fit(features_scaled, target)
    
    predictions = best_model.predict(features_scaled)
    spearman_corr, _ = spearmanr(predictions, target)
    r2 = r2_score(target, predictions)
    rmse_value = np.sqrt(mse(target, predictions))
    
    results[name] = {
        'Spearman Corr': spearman_corr,
        'R2': r2,
        'RMSE': rmse_value
    }

# Print the results
for name, metrics in results.items():
    print(f"{name}:")
    print(f"  Spearman Correlation: {metrics['Spearman Corr']:.4f}")
    print(f"  R2: {metrics['R2']:.4f}")
    print(f"  RMSE: {metrics['RMSE']:.4f}")

```

```python
from sklearn.metrics import mean_squared_error as mse, r2_score, root_mean_squared_error
from sklearn.model_selection import GridSearchCV
import pandas as pd

# Function to evaluate models
def evaluate_models(models, params, X_train, y_train, X_test, y_test):
    results = {}
    for name, model in models.items():
        print(f"Training {name}...")
        # Perform hyperparameter tuning if applicable
        if name in params:
            grid_search = GridSearchCV(model, params[name], scoring='r2', cv=5)
            grid_search.fit(X_train, y_train)
            best_model = grid_search.best_estimator_
        else:
            best_model = model
            best_model.fit(X_train, y_train)
        
        # Predictions and performance evaluation
        y_pred = best_model.predict(X_test)
        r2 = r2_score(y_test, y_pred)
        mse_score = root_mean_squared_error(y_test, y_pred)  # Using RMSE
        
        results[name] = {
            'R2': r2,
            'RMSE': mse_score,
            'Model': best_model
        }
    return results

# Train and evaluate models
results = evaluate_models(models, params, features_scaled, target, features_scaled, target)

# Sort models by R² score and get the best model
best_model_name, best_model_metrics = max(results.items(), key=lambda item: item[1]['R2'])
best_model = results[best_model_name]['Model']

print(f"Best model: {best_model_name}")

# Function to print feature importances
def print_feature_importances(model, name):
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        feature_importances = pd.DataFrame({'Feature': features.columns, 'Importance': importances})
        feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
        print(f"Feature Importances for {name}:")
        print(feature_importances)
    else:
        print(f"Feature importance is not available for {name}.")

# Print feature importances for the best model
print_feature_importances(best_model, best_model_name)
```
